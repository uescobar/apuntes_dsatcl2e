# Configuracion de ambiente local

La configuracion del entorno de trabajo para trabajar los ejemplos del libro se muestra a continuacion.

Nota: No se configura Apache Spark.

# Configurar ambiente con instrucciones del libro

Una vez instalado el entorno bajando docker

Bajamos los ejemplos de la siguiente liga:

<a href="docs/ch00/data.zip" download>data.zip</a>

Descomprimir el archivo.

Dentro del directorio de datos que ofrece el autor, ejecutamos el entorno con 

`docker run --rm -it -v "$(pwd)":/data datasciencetoolbox/dsatcl2e`
